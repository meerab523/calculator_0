```python
#calculator
num_calculations = int(input("How many calculations do you want to perform? \n"))

if num_calculations > 3:
    print("Sorry! You can perform maximum 3 calculations only.")
    num_calculations = 3

for i in range(num_calculations):
    print(f"\nCalculation {i+1}:")

    try:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
    except ValueError:
        print("Error: Please enter valid numbers!")
        continue

    operator = input("Enter an operator (+, -, *, /): ")

    if operator == '+':
        result = num1 + num2
        print(f"Result: {num1} + {num2} = {result}")
    elif operator == '-':
        result = num1 - num2
        print(f"Result: {num1} - {num2} = {result}")
    elif operator == '*':
        result = num1 * num2
        print(f"Result: {num1} * {num2} = {result}")
    elif operator == '/':
        if num2 == 0:
            print("Error: Division by zero is not allowed!")
        else:
            result = num1 / num2
            print(f"Result: {num1} / {num2} = {result}")
    else:
        print("Error: Invalid operator!")
```
